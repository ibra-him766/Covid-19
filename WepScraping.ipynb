```python
def removeNo(city):
    
    city=re.sub("\d+", "", city)
    city=re.sub(r'[^\w]', ' ', city)
    city=re.sub("\s\s+" , " ", city)
    return city.strip() 
```

```python
prevouis["city"]=prevouis["city"].apply(lambda x: removeNo(str(x)))
```

```python
def mapPlot(features):
    if features == "active":
        color="rgb(161, 50, 0)"
    elif features == "confirmed":
        color="rgb(161, 50, 0)"
    
    elif features == "recovered":
        color="rgb(95, 145, 255)"
        
    else:
        color="rgb(124,124,124)"
            
    fig = px.scatter_geo(prevouis,lat="lat",lon="lon",color='city',color_discrete_sequence=[color],
                     hover_name="city", size=features, projection="natural earth")
    fig.update_layout(    geo = dict(
            coastlinewidth = 2,
        landcolor = 'rgb(240, 240, 245)',
            lataxis = dict(
                range = [10, 35],
                showgrid = True,
                dtick = 10
            ),
            lonaxis = dict(
                range = [25, 65],
                showgrid = True,
                dtick = 20
            ),
        ))
    fig.show()

```

```python

import ipywidgets as widgets
from IPython.display import display 
b=widgets.ToggleButtons(options=["active",'confirmed','recovered','deaths'],description="No . of")


widgets.interact(mapPlot,features=b)
```
